<b>Given:</b>

The <b>Null Hypothesis H0</b> is <b>"Has the Apparent temperature and humidity compared monthly across 10 years of the data indicate an increase due to Global warming".</b>

The <b>Alternative Hypothesis H1</b> is <b>"Has the Apparent temperature and humidity compared monthly across 10 years of the data not indicate an increase due to Global warming"</b>

```python
montly_data
```

```python
#Creating new DataFrame only for Apparent Temperature and Humidity
montly_data_mean= montly_data[['Apparent Temperature (C)','Humidity']]
montly_data_mean
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(montly_data_mean, height=8).map(plt.scatter, "Apparent Temperature (C)", "Humidity")
plt.show()
```

<b>
    Observation:
</b>
There can be or may be Linear Relationship between "Humidity" and "Apparent Temperature (C)" with negative slope.

```python
plt.figure(figsize=(14,6))
sns.lineplot(data = montly_data_mean)
plt.show()
```

<b>Observation:
</b>
"Humidity" is remain constant from 2006 - 2016
But "Apparent Temperature (C)" is frequently changes from 2006 - 2016.

## 2D Scatter Plot with Color Coding for each Summary type.


```python
sns.set_style("whitegrid")
sns.FacetGrid(data_target, hue="Summary", height=8).map(plt.scatter, "Apparent Temperature (C)", "Humidity").add_legend()
plt.show()
```

<b>Observation:
</b>
    There are very few outlier.
   
   Mostly Weather is <b>Clear</b> or <b>Partly Cloudy/Rain</b> in Finland.

   Only few days there has a <b>Light Rain</b> or <b>Dry or Dangerously Windy and Partly Cloudy.</b>

### <b>Univariate Analysis using PDF:</b>

```python
# For Apparent Temperature (C)
sns.FacetGrid(data_target, hue="Summary", height=10).map(sns.distplot, "Apparent Temperature (C)").add_legend()
plt.show()
```

```python
# For Humidity
sns.FacetGrid(data_target, hue="Summary", height=10).map(sns.distplot, "Humidity").add_legend()
plt.show()
```

<b>Observation:</b>

"Humidity" is better Feature than "Apparent Temperature (C)".

## <b>Conclusion:</b>
H0 is not accepted because there is no change in Humidity from 2006–2016. So, we will accept the H1.
